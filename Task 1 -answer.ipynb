### Step 0
Load the energy data from the file `Energy Indicators.xls`, which is a list of indicators of [energy supply and renewable electricity production](Energy%20Indicators.xls) from the [United Nations](http://unstats.un.org/unsd/environment/excel_file_tables/2013/Energy%20Indicators.xls) for the year 2013, and should be put into a DataFrame with the variable name of **energy**.

Keep in mind that this is an Excel file, and not a comma separated values file. Also, make sure to exclude the footer and header information from the datafile. The first two columns are unneccessary, so you should get rid of them, and you should change the column labels so that the columns are:

`['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']`

Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule). For all countries which have missing data (e.g. data with "...") make sure this is reflected as `np.NaN` values.

Rename the following list of countries (for use in later questions):

```"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"```

There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, 

e.g. 

`'Bolivia (Plurinational State of)'` should be `'Bolivia'`, 

`'Switzerland17'` should be `'Switzerland'`.

````python
# importing numby 
import numpy as np
#importing pandas
import pandas as pd
#1 importing df_energy 
energy = pd.read_excel('data/Energy Indicators.xls',
                      header=9,
                      skiprows=[10,11,12,13,14,15,16,17],
                      skipfooter =38,           #footer=283-245=38
                      usecols="C:F",
                      na_values="...")   
````

````python
energy.head()
````

````python
energy.shape
````

````python
#remove any rows that have any duplicates
energy.drop_duplicates(inplace=True)
# check the shape
energy.shape
````

````python
#2
#rename column "Renewable Electricity Production"
energy.rename(columns={'Renewable Electricity Production' : '%Renewable'},inplace=True)
# check the new column names
energy.columns
````

````python
#Convert Energy Supply to gigajoules (there are 1,000,000 gigajoules in a petajoule)
energy['Energy Supply'] = 1000000*energy['Energy Supply']
````

````python
#3 removing parentheses
energy['Country'] = energy['Country'].str.replace(" \(.*\)","")
  
# 'Bolivia (Plurinational State of)' should be 'Bolivia',
energy.loc[energy['Country']=='Bolivia'] 
````

````python
#4
def remove_digits(text):
    clean_text =''.join([item for item in text if not item.isdigit()])   
    return clean_text


#apply the function to the country column
energy['Country'] = energy['Country'].apply(remove_digits)

#Rename the following list of countries (for use in later questions):
energy["Country"].replace({"Republic of Korea": "South Korea", "United States of America": "United States",
                           "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                          "China, Hong Kong Special Administrative Region": "Hong Kong"},inplace=True)

#'Switzerland17' should be 'Switzerland'. 
options = ['United Kingdom', 'Switzerland'] 
energy.loc[energy["Country"].isin(options)]
````

Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```python
ContinentDict  = {'China':'Asia',

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
join_result['population']=join_result['Energy Supply']/ join_result['Energy Supply per capita']  
join_result[['population']]
````

````python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

group=join_result['population'].groupby(ContinentDict)

size = group.size()
sum = group.sum()
mean = group.mean()
std = group.std()
result = pd.DataFrame({
        'size': size,
        'sum': sum,
        'mean': mean,
        'std': std})
result
````
